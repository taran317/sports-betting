```python
game_dates = []
for date in df_games['game_date']:
    splitted = date.split("-")
    game_dates.append(splitted[1] + "/" + splitted[2] + "/" + splitted[0])
game_home_teams = []
for matchup in df_games['matchup']:
    if 'vs.' in matchup:
        game_home_teams.append(matchup.split(" vs. ")[0])
    else:
        game_home_teams.append(matchup.split(" @ ")[1])
dates_teams = set([(game_dates[i] + game_home_teams[i]) for i in range(len(game_dates))])
```
